```python
# Remove All The Marked Elements of a List
# [1, 1, 2, 3, 1, 2, 3, 4], [1, 3] -> [2, 2, 4]
# [1, 1, 2, 3, 1, 2, 3, 4, 4, 3, 5, 6, 7, 2, 8], [1, 3, 4, 2] -> [5, 6, 7, 8]
# [8, 2, 7, 2, 3, 4, 6, 5, 4, 4, 1, 2, 3], [2, 4, 3] -> [8, 7, 6, 5, 1]

def rmv_marked(lst1, lst2):
    for j in lst1[:]:
        for i in lst2[:]:
            if i == j:
                lst1.remove(i)
    return lst1
```

```python
rmv_marked([1, 1, 2, 3, 1, 2, 3, 4], [1, 3])
```

```python
rmv_marked([1, 1, 2, 3, 1, 2, 3, 4, 4, 3, 5, 6, 7, 2, 8], [1, 3, 4, 2])
```

```python
rmv_marked([8, 2, 7, 2, 3, 4, 6, 5, 4, 4, 1, 2, 3], [2, 4, 3])
```

```python
# Method For Counting Total Occurence Of Specific Digits

# method of list class.
# 1 given list of int, a second given list of marked digit
#list of tuples with 2 elements: digit to count, the count

def count_spec_digits(integers_lists, digist_list):
    
    #first converting any negative number
    for i in range(0, len(integers_lists)):
        if integers_lists[i] < 0:
            integers_lists[i]*=(-1)
    print(integers_lists)
    
    #creating string list to ease the check
    str_int=[]
    str_dig=[]
    
    for i in integers_lists:
        str_int.append(str(i))
    
    #joining the list of string, otherwise it wasn't for figure about 9 in my company   
    str_int = ''.join(str_int)

    for i in digist_list:
        str_dig.append(str(i))
        
    count_lst=[]
    for i in str_dig:
        count_lst.append(str_int.count(i))

    results = [(digist_list[i], count_lst[i]) for i in range(0, len(count_lst))]  
    return results
```

```python
count_spec_digits([1, 1, 2 ,3 ,1 ,2 ,3 ,4], [1, 3])
```

```python
count_spec_digits([-18, -31, 81, -19, 111, -888],[1, 8, 4])
```

```python
count_spec_digits([-77, -65, 56, -79, 6666, 222],[1, 8, 4])
```

```python
#Count the number of occurrences of each character 
#and return it as a (list of tuples) in order 
#of appearance. For empty output return (an empty list).

#Example:

#orderedCount("abracadabra") == [['a', 5], ['b', 2], ['r', 2], ['c', 1], ['d', 1]]

def orderedCount(text):
    
    single_letter=[]

    for i in text:
        if i not in single_letter:
            single_letter.append(i)

    cnt_let=[]

    for i in single_letter:
        cnt_let.append(text.count(i))
    print(single_letter)
    print(cnt_let)

    fin_results = [[single_letter[i], cnt_let[i]] for i in range(0, len(single_letter))]  
    return fin_results
```
