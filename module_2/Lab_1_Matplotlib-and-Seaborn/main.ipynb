#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here

x = np.arange(0,100)
y = x**2
z = np.exp(x)

fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(10,10))
ax[0].set_title("square of x")
ax[0].plot(x,y, linewidth=2, linestyle="solid")
ax[0].set_xlabel('x')     
ax[0].set_ylabel('y = x^2')
ax[0].plot(x,z, linewidth=2, linestyle="solid")

ax[1].set_title("Z chart")
ax[1].plot(x,z, linewidth=2, linestyle="solid")
ax[1].set_xlabel('x')     
ax[1].set_ylabel('z = exp(x)')
ax[1].set_xscale('log')
ax[1].set_yscale('log')
ax[1].plot(x,y, linewidth=2, linestyle="solid")


plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here
x = np.arange(0,100)
y = x**2
z = np.exp(x)

fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(10,10))
ax[0].set_title("Chart of x squared")
ax[0].plot(x,y, linewidth=2, linestyle="solid", label= "Exponential value of x")
ax[0].set_xlabel('x')     
ax[0].set_ylabel('y = x^2')
ax[0].legend(loc="upper left")

ax[1].set_title("Chart of x exponentialized")
ax[1].plot(x,z, linewidth=2, linestyle="solid", label= "Square value of x")
ax[1].set_xlabel('x')     
ax[1].set_ylabel('z = exp(x)')
ax[1].set_xscale('log')
ax[1].set_yscale('log')
ax[1].legend(loc="upper left")

plt.show()
```

#### We would like to have in one single plot the summary statistics of the feature `Age`. What kind of plot would you use? Plot it. 

```python
# your code here

sns.displot(titanic, x="age", bins=80, color='grey' , kde=True)

mean = titanic.age.mean()
median = titanic.age.median()
quart_1 = titanic.age.quantile(0.25)
quart_3 = titanic.age.quantile(0.75)

plt.axvline(mean, color='r',linestyle='solid', linewidth=2)
plt.axvline(median, color='r', linestyle='dotted', linewidth=2)
plt.axvline(quart_1, color='r', linestyle='dotted', linewidth=2)
plt.axvline(quart_3, color='r', linestyle='dotted', linewidth=2)

plt.show()
```
