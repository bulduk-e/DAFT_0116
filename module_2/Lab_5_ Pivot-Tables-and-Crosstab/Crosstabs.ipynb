### 2. Use crosstabs to show the number of moovies per country

```python
movies_per_country = pd.DataFrame(pd.crosstab(index=movies.country, columns=movies.country, margins=True)["All"])
```

```python
movies_per_country
```

### 4. Average duration of filmes grouped by color

```python
duration_percolor = pd.DataFrame(pd.crosstab(movies.color, movies.color, values=movies.duration, margins=True, aggfunc='mean').round(1)["All"])
```

```python
duration_percolor
```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
movies.columns
```

```python
movies["total_fb_likes"]=movies["director_facebook_likes"]+movies["actor_3_facebook_likes"] \
    +movies["actor_1_facebook_likes"]+movies["cast_total_facebook_likes"]+movies["actor_2_facebook_likes"]+movies["movie_facebook_likes"]
```

```python
movies
```

```python
like_per_dir = pd.DataFrame(pd.crosstab(movies.director_name, movies.director_name, values=movies.total_fb_likes, margins=True, aggfunc='mean').round(0)["All"])
```
